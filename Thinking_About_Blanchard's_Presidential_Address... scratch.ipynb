# <font color="880000">Thinking About Blanchard's Presidential Address...</font>

One major conclusion from Olivier Blanchard's excellent, thoughtful, and provocative AEA Presidential Address "Public Debt and Low Interest Rates" <https://www.aeaweb.org/aea/2019conference/program/pdf/14020_paper_etZgfbDr.pdf> is that whether a marginal increase in public debt raise welfare—whether an economy is "dynamically inefficient" depends on the relationship of both rate of interest on safe government bonds and the rate of profit to the economy's growth rate. In the model he presents:

1. What he calls the "partial-equilibrium" transfer effect raises welfare if the rate on safe government bonds is less than the growth rate. 
2. What he calls the "general-equilbrium" factor-price effect working through crowding-out raises the welfare of the average agent only when the risky profit rate is less than the growth rate.

I may be being stupid here—my estimate of the chance I am being stupid is about 75%. But when I tried to work through Blanchard's algebra for the Constant-Absolute-Risk-Aversion—CARA—case, I found that it was important for Blanchard's second conclusion that he analyzed, as Robert Waldmann puts it, not the average effect but rather the effect on the average. I think I have results for the average effect in the CARA case that are different from Blanchard's. While the average CARA agent benefits from the general-equilibrium crowding-out factor-price effect only if the risky rate of profit is lower than the growth rate, the average effect is a benefit if not the risky rate of profit, not the safe rate of interest, but rather a third shadow rate of return which is certainly lower than the risky profit rate and may be higher or lower than the safe rate is lower than the economy's growth rate.

&nbsp;

## <font color="880000">My Deep, Conceptual Problem...</font>

I understand how the general equilibrium-effect gain when you are old from higher profits produced by crowding-out is correlated with consumption, and hence is worth less then the product of its expected value and the expected value of marginal utility. I get how it is thus appropriately discounted not at the risk-free but at the risky rate.

But here is the rub: the general equilibrium-effect reduction in your wage when you are young is also correlated with consumption. So the certain-equivalent cost value of what the general equilibrium-effect makes you give up when young is also less than its expected value times your expected marginal utility.

With a risk-free cost and a risk-free benefit, the partial-equilibrium effect of a transfer program is properly assessed using the risk-free rate.

With a risk-free cost and a risky benefit, the general-equilibrium effect of a transfer program would be properly assessed using the risky rate.

But in the general-equilibrium effect we have a risky cost and a risky benefit—and I cannot see how the risky rate is then appropriate.

Thus I cannot see where the conclusion that it is appropriate to discount the benefits from the general-equilibrium effect at the risky rate comes from.

## <font color="880000">I Think I Have Made Progress, but I Am Probably Wrong...</font>

Let me see if I can get this exactly right:

Blanchard‘s extremely interesting, provocative, and masterly AEA presidential address concludes that the “general equilibrium“ effect of debt working through crowding-out and factor prices should be assessed at the average profit rate rather than the risk-free government bond rate. Because the value of the higher profit rate is correlated with consumption, it has a lower utility value and needs to be discounted more—hence discounted by the risky profit rate (remember: consumption when old is just profits, and so they are 100% correlated) when the value of higher profits when old is compared to the cost of lower wages when young.

What Blanchard has, I think, missed is that in the uncertainy case wages received are also risky, and in the CARA case as risky as profits. The same discount from their expected value it needs to be applied to them. And when we apply this risk discount to both terms of the general equilibrium perfect and when we apply this risk discount to both terms of the general equilibrium effect— to both higher profits and lower wages— we get back to the conclusion that the right intertemporal discount rate to apply is the risk-free rate.

Note that this is for the CARA case only...

### <font color="000088">Certainty Case</font>

>(1) $ U = U_y + {\beta}U_o $ :: Utility

>(2) $ C_y = W_y - K_o - D ; C_o = RK_o + D $ :: Budget Constraint

>(3) $ Y = F(K,1) $ :: Production Function

>(4) $ R = \frac{dF(K, 1)}{dK} $ :: Rate of Profit

>(5) $ U'(C_y) = {\beta}RU'(C_o) $ :: First-Order Condition

>(6) $ \frac{dU}{dD} = \frac{dU_a}{dD} + \frac{dU_b}{dD} $ :: Separate Out Partial- and General-Equilibrium Effects

>(7) $ \frac{dU_a}{dD} = -U'(C_y) + {\beta}U'(C_o) $ :: Partial-Equilibrium Effect

>(8) $ \frac{dU_b}{dD} = U'(C_y)\frac{dW}{dD} +  {\beta}U'(C_o)K\frac{dR}{dD} $ :: General Equilibrium Effect

Using the first-order condition (4) to evaluate $ U_a $:

>(9) $ \frac{dU_a}{dD} = -{\beta}RU'(C_o) + {\beta}U'(C_o) $

>(10) $ \frac{dU_a}{dD} = {\beta}(1-R)U'(C_o) $

Hence you evaluate costs and benefits of the partial-equilibrium effect using the profit rate

Now turning to the general-equilibrium effect:

>(11) $ \frac{dW}{dR} = -K $ :: Factor-Price Frontier

>(12) $ \frac{dU_b}{dD} = \left[ -U'(C_y)K +  {\beta}U'(C_o)K \right] \frac{dR}{dD} $

>(13) $ \frac{dU_b}{dD} = \left[ -{\beta}RU'(C_o)K +  {\beta}U'(C_o)K \right] \frac{dR}{dD} $

>(14) $ \frac{dU_b}{dD} = (1-R)U'(C_o)K\frac{dR}{dD} $

Hence you evaluate costs and benefits of the general-equilibrium effect using the profit rate

### <font color="000088">Uncertainty Case</font>

Uncertainty from a productivity shock:

>(15) $ Y_t = (1+\epsilon_t) F(K_t, 1) $

First-order Condition:

>(16) $ U'(C_{y,t}) = {\beta}E\left[ R_{t+1}(1+\epsilon_{t+1})U'(C_{o,t+1}) \right]   $

Shadow risk-free rate:

>(17) $ {R^f}_{t+1}E[U'(C_{o,t+1})] = E[R_{t+1}(1+\epsilon_{t+1})U'(C_{o,t+1})] $

The change in utility:

>(18) $ \frac{dU_t}{dD} = \frac{dU_{a,t}}{dD} + \frac{dU_{b,t}}{dD} $

>(19) $ \frac{dU_{a,t}}{dD} = -E[U'(C_{y,t})] + E[{\beta}U'(C_{o,t+1})] $

>(20) $ \frac{dU_{b,t}}{dD} = E\left[U'(C_{y,t})\frac{dW_t}{dD}\right] +  E\left[{\beta}U'(C_{o,t+1})K_{t+1}(1+\epsilon_{t+1})\frac{dR_{t+1}}{dD}\right] $

&nbsp;

#### <font color="008800">The Partial-Equilibrium Term</font>

Substituting the risk-free rate definition (17) into the first order condition (16) and then into (19) to assess the partial-equilibrium welfare term:

>(21) $ \frac{dU_{a,t}}{dD} = -{\beta}\left[{R^f}_{t+1}E[U'(C_{o,t+1})]\right] + {\beta}E[U'(C_{o,t+1})] $

>(22) $ \frac{dU_{a,t}}{dD} = {\beta}(1-{R^f}_{t+1})E\left[E[U'(C_{o,t+1})]\right] $

The right interest rate to use for the partial-equilibrium term is the risk-free rate $ {R^f}_{t+1} $.

&nbsp;

#### <font color="008800">The General-Equilibrium Term</font>

Use the factor-price frontier relationship:

>(23) $ \frac{dW_t}{dR_t} = (1+\epsilon_t)K_t $

To somewhat simplify (20):

>(24) $ \frac{dU_{b,t}}{dD} = -E\left[U'(C_{y,t})(1+\epsilon_{t})K_t\frac{dR_t}{dD}\right] +  E\left[ {\beta}U'(C_{o,t+1})K_{t+1}(1+\epsilon_{t+1})\frac{dR_{t+1}}{dD} \right] $

>(25) $ \frac{dU_{b,t}}{dD} = -E\left[U'(C_{y,t})(1+\epsilon_{t}) \right] K_t \frac{dR_t}{dD} + 
E\left[ {\beta}U'(C_{o,t+1})(1+\epsilon_{t+1}) \right] K_{t+1} \frac{dR_{t+1}}{dD} $

&nbsp;

#### <font color="008800">Specialize to the CARA Case</font>

There are the two shocks $ \epsilon_t $ and $ \epsilon_{t+1} $. There is one persistent dimension of the state of the system as of the start of time t, determined and indexed by $ K_t $. Since I do not know the distribution of $ K_t $, I do not want to carry around that state—it makes taking expected utility impossible.

If we go CARA, then investments in risky assets depend only on their risk and return and not on wealth our persistent state disappears: $ K_t = K_{t+1} = K $, $ R_t = R_{t+1} = R $, our first-order condition (16) becomes:

>(26) $ U'(C_{y,t}) = {\beta} R E\left[(1+\epsilon_{t+1})U'(C_{o,t+1}) \right] $

and (25) becomes:

>(27) $ \frac{dU_{b}}{dD} = \left[ -E\left[U'(C_{y,t})(1+\epsilon_{t}) \right] + {\beta}
E\left[U'(C_{o,t+1})(1+\epsilon_{t+1}) \right] \right] K \frac{dR}{dD} $

We can then divide through by the period-t certainty-equivalent utility value of a unit of extra profits:

>(28) $ \frac{\frac{dU_{b}}{dD}}{{\beta} E\left[ U'(C_{o,t+1})(1+\epsilon_{t+1}) \right]]} = \left[1 - \frac{E\left[U'(C_{y,t})(1+\epsilon_{t}) \right] }{ {\beta} E\left[ U'(C_{o,t+1})(1+\epsilon_{t+1}) \right] } \right] K \frac{dR}{dD} $

And then sustitute in the first-order condition:

>(29) $ \frac{\frac{dU_{b}}{dD}}{{\beta} E\left[ U'(C_{o,t+1})(1+\epsilon_{t+1}) \right]]} = \left[1 - R \left( 1 + \frac{ E\left[ U'(C_{y,t})\epsilon_{t} \right] }{ E\left[U'(C_{y,t}) \right] } \right) \right] K \frac{dR}{dD} $

I think that this (29) is what we are looking for. 

The right discount factor to apply to the higher future profits generated by crowding out is not the average risky profit rate $ R $ but instead:

>(30) $ R^{U_b} = R \left( 1 + \frac{ E\left[ U'(C_{y,t})\epsilon_{t} \right] }{ E\left[U'(C_{y,t}) \right] } \right) $

Since the marginal utility of consumption when young $ C_{y,t} $ is negatively correlated with the period-t productivity shock $ \epsilon_t $, this $ R^{U_b} $ will be lower than $ R $. How much lower? Well, we can rewrite our shadow risky and risk-free rate expression (17) to get the risk-free rate relative to the risky rate:

>(31) $ {R^f} = R \left(1 + \frac{E[U'(C_{o,t+1})\epsilon_{t+1}]}{E[U'(C_{o,t+1})]} \right) $

And it looks to be a similar thing...

Why is it a similar thing? I think it is the same thing because CARA also forces consumption when young to be correlated with wages in the same way that consumption when old is correlated with profits. Thus you have to apply a similar discount to get the certain-equivalent value when young of the wage cost as to get the certain-equivalent value when old of the profit benefit.

&nbsp;

#### <font color="008800">CARA Case Conclusion</font>

The right interest rate to use for the general-equilibrium term is neither the expected profit rate $ R $ nor the risk-free rate $ {R^f}_{t+1} $, but instead $ R^{U_B} $, which is certainly lower than $ R $, and could be higher or could be lower than $ R^f $, depending. Depending on what? Depending on whether the negative covariance of the current shock with marginal utility divided by expected marginal utility is when the agent is young (in which case $ R^{U_b} $ will be less than $ R^f $; the general-equilibrium effect is then a negative-beta "insurance" asset) or when the agent is old (in which case $ R^{U_b} $ will be greater than $ R^f $; the general-equilibrium effect is then a positive-beta "risky" asset).
