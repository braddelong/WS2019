# <font color="880000">Venus</font>

OK. Venus is at its maximum angular separation from the Sun when Earth-Venus-Sun [form a right triangle](https://en.wikipedia.org/wiki/Elongation)—which is when the Earth-Venus line-of-sight is tangent to the orbit of Venus... 

&nbsp;

```python
%matplotlib inline

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
import pandas as pd
```

```python
theta = []
thetaprime = []
thetadiff = []

b = 0.723
pi = 3.141592653

for i in range(-500,500):
    thetai = pi*i/1000
    y = np.sin(thetai)
    x = np.cos(thetai)
    yprime = y - b
    xprime = x
    thetaprimei = np.arctan(yprime/xprime)
    thetadiffi = thetai - thetaprimei
    theta = theta + [thetai]
    thetaprime = thetaprime + [thetaprimei]
    thetadiff = thetadiff + [thetadiffi]

data = [('theta', theta),('thetaprime', thetaprime), ('thetadiff', thetadiff)]
venus = pd.DataFrame.from_items(data)
venus.plot()
```
